```python
import pandas as pd 
```

```python
df=pd.read_csv("Datasets/insurance_dataset.csv")
```

```python
df.describe()
```

```python
#This list will store one text record per row
#Each record = one natural-language description of a customer
records =[]
```

```python
#df.iterrows() goes row by row
for i, row in df.iterrows():
#""" → multi-line string (easy to read,clean formatting)
    record=f"""
insurance record {i+1}

This insurance record describes a customer aged {row['Age']}.
The customer is {row['Gender']} and works as a {row['Occupation']}.
Their education level is {row['Education']} and marital status is {row['Marital_Status']}.
The customer has an annual income of {row['Income']}.
The recorded insurance claim amount for this profile is {row['Claim_Amount']}.
"""
#.strip() removes extra blank lines at the start and end
    records.append(record.strip())
#Joining all records into one document
final_text = "\n\n---\n\n".join(records)
#Writing the text to a file
with open("insurance_claims_records.txt", "w", encoding="utf-8") as f:
    f.write(final_text)

print("insurance_claims_records.txt created successfully")
print("Total records created:", len(records))
```
