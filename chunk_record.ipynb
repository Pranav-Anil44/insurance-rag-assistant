```python
from pathlib import Path
#pathlib is a modern Python library for working with files and folders
#it is for easier file operations
```

```python
file_path=Path("insurance_claims_records.txt")
#This creates a Path object pointing to:insurance_claims_recprds.txt
```

```python
text = file_path.read_text(encoding="utf-8")
```

```python
#Split using separator
#Some elements may contain extra spaces or newlines #Some elements may be empty strings ("") we don't want these
raw_chunks=text.split("---")

#Clean chunks #.strip() removes: Leading spaces Trailing spaces ,Leading/trailing newlines
#Only keep this chunk if it contains actual text.that is why we use if chunk.strip()
chunks= [chunk.strip() for chunk in raw_chunks if chunk.strip()]

#Verify 
print("Total chunks created:", len(chunks))

#inspect first and last chunk for sanity.
print("\nFirst chunk preview:\n")
print(chunks[0][:300])

print("\nLast chunk preview:\n")
print(chunks[-2][:300])
```
