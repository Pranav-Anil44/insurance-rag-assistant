```python
from pathlib import Path
```

```python
# We must confirm:The raw knowledge source is readable and intact.This step simply loads the full text file into memory
#This does NOT read the file.It just creates a reference like
file_path=Path("insurance_claims_records.txt")


text=file_path.read_text(encoding="utf-8")

print("file loaded successfully")
print("Total charecters:",len(text))
```

```python
raw_chunks = text.split("---")
chunks = [chunk.strip() for chunk in raw_chunks if chunk.strip()]

print("Total chunks:", len(chunks))
```

```python
print(type(chunks))
print(type(chunks[0]))
```

```python
#are loading a pretrained transformer model that converts text into vectors.
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded successfully")
```

```python
#generate one  test embedding
#We are passing one insurance record to the model.
#Internally the model:Tokenizes the text,Passes tokens through transformer layers,Extracts sentence embedding
#Returns a numeric vector
test_embedding=model.encode(chunks[0])

#The model all-MiniLM-L6-v2 outputs:384-dimensional embeddings.
#One insurance record converted into 384 numbers
#Those numbers are not random.They represent:Age patterns,Income patterns,Claim amount relationships,Text structure,Semantic meaning
print("Single embedding shape: ", test_embedding.shape)


print("First 10 values", test_embedding[:10])
```

```python
#Now we are scaling the model to 13000 records
embeddings=model.encode(chunks,show_progress_bar=False)
print("Embedding complete")
print("Shape:", embeddings.shape)
```

```python
import numpy as np 
embeddings=embeddings.astype("float32")
np.save("insurance_embedding.npy",embeddings)
print("Embeddings saved successfully")
print("Saved shape:",embeddings.shape)
```
